```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Load the tomato dataset
df = pd.read_csv('/content/drive/MyDrive/Proyecto final/Tomato.csv', parse_dates=['Date'])
df = df.set_index('Date')


# Create the lag features
lags = [1, 7, 30, 90]

for lag in lags:
    df[f'avg_price_{lag}'] = df['Average'].shift(lag)

# Remove rows with missing values
df = df.dropna()

# Split the data into training and testing sets
split_date = '2020-01-01'
train = df.loc[df.index < split_date]
test = df.loc[df.index >= split_date]
# Fit a linear regression model on the training data
features = ['avg_price_1', 'avg_price_7', 'avg_price_30', 'avg_price_90']
target = 'Average'
model = LinearRegression()
model.fit(train[features], train[target])

# Predict on the testing data
test['predicted_price'] = model.predict(test[features])


# Calculate the MSE and RMSE
mse = mean_squared_error(test[target], test['predicted_price'])
rmse = np.sqrt(mse)

# Print the RMSE
print(f'RMSE: {rmse:.2f}')

# Visualize the predicted vs. actual prices using Plotly Express
fig = px.line(test, x=test.index, y=[target, 'predicted_price'], title='Tomato Prices')
fig.show()
```
